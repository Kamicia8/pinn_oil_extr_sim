```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
metrics = pd.read_csv("/mnt/c/PINN_mgr/FEniCS/eq_newton/results/metrics/metrics.csv")
```

```python
print(metrics.head())
```

```python
plt.figure(figsize=(6,4))
plt.plot(metrics['time'], metrics['delta_L2'], marker='o')
plt.xlabel("Czas")
plt.ylabel("ΔL2 względem poprzedniego kroku")
plt.title("Zmiana pola w czasie")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(metrics['time'], metrics['max_u'], marker='o', color='r')
plt.xlabel("Czas")
plt.ylabel("max(u)")
plt.title("Maksymalna wartość pola w czasie")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(metrics['time'], metrics['mean_u'], marker='o', color='g')
plt.xlabel("Czas")
plt.ylabel("Średnia wartość pola")
plt.title("Średnia wartość pola w czasie")
plt.grid(True)
plt.show()
```
